```python
#outlier treatment
import seaborn as sns
sns.boxplot(x=census['instance weight'])
```

```python
# Normal formulae for interquartile range
Q1 = census['instance weight'].quantile(0.25)
Q3 = census['instance weight'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
limit = 1.5 * IQR
floor = limit - Q1 
cap = Q3 + limit
print(floor)
print(cap)
census.quantile([1,0.02,0.09,1])
```

```python
census_out = census[(census['instance weight']>=floor)&(census['instance weight']<=cap)]
```

```python
sns.boxplot(x= census_out['instance weight'])
```

```python
#z-score
int_columns =  census.select_dtypes(['int64']).columns
int_columns.value_counts()
```

```python
from scipy.stats import zscore
census_z = census[int_columns].apply(zscore)
census_z.head()
```

```python
#converting categorical data to numeric data
def cat_numeric(x):
    if x==' Female':
        return 0
    if x==' Male':
        return 1
census['gender_cd']=census['gender'].apply(cat_numeric)
census.head()
```

```python
#one-hot encoding
dummies = pd.get_dummies(census['maritial status'])
dummies.head()
```

```python
census_dummies = pd.get_dummies(census)
census_dummies.head()
```

```python
obj_columns = census.select_dtypes(['object']).columns
for col in obj_columns:
    census[col] = census[col].astype('category')
```
